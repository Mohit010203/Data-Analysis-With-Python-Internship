## Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load dataset

```python
# Use pandas to read student-mat.csv dataset
data = pd.read_csv("student-mat.csv", sep=';')
data.columns = data.columns.str.strip()
data.head()
```

## Explore & Clean Data

```python
#Check missing values (.isnull())
data.isnull().sum()
```

```python
#Remove duplicates
data = data.drop_duplicates()
data.shape
```

```python
#Inspect dataset shape & dtypes.
print("\nShape:", data.shape)
print("\nData Types:")
print(data.dtypes)
```

## Analysis Questions

```python
# Average final grade (G3).
avg_grade = data['G3'].mean()
print("Average Final Grade:", avg_grade)
```

```python
# How many students scored above 15?
above_15 = data[data['G3'] > 15].shape[0]
print("Students scoring above 15:", above_15)
```

```python
# Is study time correlated with performance?
correlation = data['studytime'].corr(data['G3'])
print("Correlation (Study time vs G3):", correlation)
```

```python
# Which gender performs better on average?
gender_avg = data.groupby('sex')['G3'].mean()
print("\nAverage score by gender:")
print(gender_avg)
```

## Visualizations

```python
# Histogram of grades.
plt.figure()
plt.hist(data['G3'], bins=10)
plt.title("Histogram of Final Grades")
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")
plt.show()
```

```python
# Scatterplot: study time vs grades.
plt.figure()
plt.scatter(data['studytime'], data['G3'])
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.show()
```

```python
# Bar chart: male vs female average score.
plt.figure()
gender_avg.plot(kind='bar')
plt.title("Average Score: Male vs Female")
plt.xlabel("Gender")
plt.ylabel("Average Final Grade")
plt.show()
```
